```python
import numpy as np
import pandas as pd
df= pd.read_csv("bweight.csv")
df.head()
```

```python
#df = df.columns['age', ]
df.shape
```

```python
#data exploration/visualisation
#the following histogram shows the distribution of age group considered in the study. 
(df['gestage']
 .value_counts(sort=False)
 .plot.barh()
 .set_title('Anything')
)
```

```python
# the following scatter plot shows the relationship between body weight and age. We can clearly see the linear relationship between them i.e. increase in weight shows incerase in age. 
df.plot(x='bweight', y='gestage', kind='scatter')
# the box plot shows the distribution of body weight acros a specific age group. This also helps in detecting the outliers.
df.boxplot(column='bweight', by='gestage')
```

```python
from matplotlib.pyplot import pie, axis, show
%matplotlib inline

sums = df.groupby(df['sex'])
axis('equal');
pie(sums, labels=sums.index);
show()
```

```python
import matplotlib.pyplot as plt
country_data = df["bweight"]
medal_data = df["gestage"]
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#8c564b", "#d62728","#d62728","#d62728"]
explode = (0, 0, 0, 0,0,0,0,0)  
plt.pie(medal_data, labels=country_data, explode=explode, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("Gold medal achievements of five most successful\n"+"countries in 2016 Summer Olympics")
plt.show()
```

```python
#violin plot showing the distribution of body weight wrt sex. Irt is not diffcult to which plot is meant for female and which one is for male.
import seaborn as sns
sns.violinplot(x='sex', y='bweight', data=df)
```
