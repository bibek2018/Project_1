# Description of data:
This data contains information of body weight, age and gender of patients coming to a hospital during certain period.

#### The above plot shows that the data is pretty evenly distributed. It is in fact more normally distributed though the number of patients is not high enough to conclude such. 

```python
# the following scatter plot shows the relationship between body weight and age. 
df.plot(x='bweight', y='gestage', kind='scatter')
# the box plot shows the distribution of body weight acros a specific age group.
df.boxplot(column='bweight', by='gestage')
```

### Salient points about above plots:
We can clearly see the linear relationship between the age and weight i.e. increase in weight shows incerase in age(which of course is not very surprising).
